```python
main_box =soup.find_all("div",{"class":"_2kHMtA"})
```

```python
data_list = []
for box in main_box:
    temp_dict={}
    temp_dict['TITLE']=box.find("div",{"class":"_4rR01T"}).text.strip()
    temp_dict['RATING']=box.find("div",{"class":"_3LWZlK"}).text.strip()
    temp_dict['PRICE']=box.find("div",{"class":"_30jeq3 _1_WHN1"}).text.strip()
    print(temp_dict)
    data_list.append(temp_dict)
```

```python
df=pd.DataFrame(data_list)
```
